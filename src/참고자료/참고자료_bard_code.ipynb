```python
import transformers

model = transformers.AutoModelForSequenceClassification.from_pretrained("google/electra-base-discriminator")

# Tokenize the text data
tokenizer = transformers.AutoTokenizer.from_pretrained("google/electra-base-discriminator")
inputs = tokenizer(text="The quick brown fox jumps over the lazy dog. The dog is sleeping.", return_tensors="pt")

# Classify the sequence
outputs = model(**inputs)
```

```python
outputs['logits']
```

```python
# Get the classification prediction for each sentence
sentence_classification_predictions = outputs["logits"].squeeze(-1)

# Print the classification predictions for each sentence
for i, sentence_classification_prediction in enumerate(sentence_classification_predictions):
    print(f"Sentence {i + 1}: {sentence_classification_prediction}")
```
